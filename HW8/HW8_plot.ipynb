```python
import matplotlib.pyplot as plt
import seaborn
import pandas as pd
```

```python
cars_data = pd.read_csv("https://raw.githubusercontent.com/umerudel/DSPS_uZubair/master/HW8/motor_trend_cars.csv")
```

```python
cars_data.head()
```

```python
cars_data.describe()
```

```python
plt.figure(figsize=(12,10), dpi= 200)
seaborn.heatmap(cars_data.corr(), xticklabels = cars_data.corr().columns, yticklabels = cars_data.corr().columns, cmap='RdYlGn', center = 0, annot = True)

plt.title('Correlogram of Motor Trend Cars', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
